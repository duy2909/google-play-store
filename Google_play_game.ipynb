**Column description:**
1. App - Application name
2. Category - Category the app belongs to
3. Rating - User rating of the app (0-5*)
4. Review - Number of review for the app
5. Size - Size of the app
6. Installs - Number of user downloads for the app
7. Type - Paid/Free
8. Price - Price of the app
9. Content rating - Age group the app target at
10.  Genres - App can belongs to multiple type
11. Last Updated - Date when the app was last updated on Play Store
12. Current Ver - Current version of the app available on Play Store
13. Android Ver - Min required Android *version*


**Data Cleaning**

```python
# Transform size to numerical data
data['Size'].unique()
```

```python
def transform_size(data):
  if 'M' in data:
    return float(data[:-1]) * 1024
  elif 'k' in data:
    return float(data[:-1])
  return data
data['Size'] = data['Size'].apply(transform_size)
```

```python
# Check invalid data
invalid_data = []
for val in data['Size']:
  if not isinstance(val, float):
    if val not in invalid_data:
      invalid_data.append(val)
data = data[~data['Size'].isin(invalid_data)]
```

```python
# Check rating
data['Rating'].unique()
```

```python
# Transform installs column
data['Installs'].unique()
```

```python
def transform_install(data):
  if '+' in data:
    return int(data[:-1].replace(',',''))
  return int(data.replace(',',''))
data['Installs'] = data['Installs'].apply(transform_install)
```

```python
# Transform reviews column
data['Reviews'] = data['Reviews'].apply(lambda x: int(x))
```

```python
# Transform price column
data['Price'].unique()
```

```python
def transform_price(data):
  if '$' in data:
    return float(data[1:].replace(',',''))
  return float(data)
data['Price'] = data['Price'].apply(transform_price)
data['Price'].unique()
```

```python
# Check category
data['Category'].unique()
# Get data which has 'Game' category
game_data = data[data['Category'] == 'GAME']
```

```python
import plotly.express as px
# Show top genres game by install
genres_install = game_data.groupby(['Genres'])['Installs'].sum().reset_index().sort_values(by='Installs', ascending=False)
genres_chart = px.bar(genres_install, x='Genres', y='Installs', title='Total Install by Genres')
genres_chart.show()
```

```python
# Check correlation between columns
game_data_corr = game_data.corr()
corr = px.imshow(game_data_corr,
                 x=game_data_corr.columns,
                 y=game_data_corr.columns)
corr.show()
```

```python
rev_ins = px.scatter(game_data, x='Reviews', y='Installs')
rev_ins.show()
```

```python
# Find different between Free and Paid app
paid = game_data.groupby(['Type'])['App'].count().reset_index()
paid
```

```python
# Explore Content Rating and install
content_rating = game_data.groupby('Content Rating')['App'].count().reset_index().sort_values(by='App', ascending=False)
content_chart = px.bar(content_rating, x='Content Rating', y='App', title='Total App by Content Rating')
content_chart.show()
```

```python
genres_content_rating = game_data.groupby(['Genres', 'Content Rating'])['App'].count().reset_index().sort_values(by='App', ascending=False)
genres_content_rating_chart = px.bar(genres_content_rating,
                                     x='Genres', y='App', color='Content Rating',
                                     title='Total App By Content Rating and Genre')
genres_content_rating_chart.show()
```
